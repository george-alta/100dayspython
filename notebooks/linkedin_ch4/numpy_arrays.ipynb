```python
sinx = np.sin(x)
```

```python
pp.plot(x, sinx)
```

```python
pp.plot(x, sinx)
```

```python
pp.plot(x, sinx)
pp.plot(x, np.cos(x))
pp.plot(x, np.log(1 + x))
```

```python
pp.plot(x, sinx, label="sin (x)")
pp.plot(x, np.cos(x), label="cos (x)")
pp.plot(x, np.log(1 + x), label="log (1+x)")

pp.legend()
```

```python
cosx = np.cos(x)
```

```python
y = sinx * cosx
z = cosx**2 - sinx**2

pp.plot(x, y)
pp.plot(x, z)
```

```python
monalisa_bw = np.loadtxt("./monalisa.txt")
```

```python
monalisa_bw.shape
```

```python
xgrad = np.linspace(0, 1, 134)
```

```python
xgrad.shape
```

```python
monalisa_xgrad = monalisa_bw * xgrad
```

```python
pp.figure(figsize=(16, 10))
pp.subplot(1, 2, 1)
pp.imshow(monalisa_bw, cmap="gray")
pp.title(label="left: normal")
pp.subplot(1, 2, 2)
pp.imshow(monalisa_xgrad, cmap="gray")
pp.title(label="right: gradient")
```

```python
ygrad = np.linspace(0, 1, 200)
```

```python
ygrad.shape
```
