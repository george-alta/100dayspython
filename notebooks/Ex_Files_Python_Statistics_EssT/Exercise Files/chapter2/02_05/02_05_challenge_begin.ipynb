Standard imports


```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
data = pd.read_csv("./tb.csv")
```

```python
data.columns
```

```python
data.head()
```

```python
dmelt = data.melt(
    ["country", "year"], ['m04', 'm514', 'm014', 'm1524', 'm2534', 'm3544',
                          'm4554', 'm5564', 'm65', 'mu', 'f04', 'f514', 'f014', 'f1524', 'f2534',
                          'f3544', 'f4554', 'f5564', 'f65', 'fu'], "group", "cases"
)
dmelt.head()
```

removing al NaN


```python
dmelt.dropna(subset=["cases"], inplace=True)
dmelt.head()
```

```python
dmelt["gender"] = dmelt["group"].str[0].str.upper()
dmelt.head()
```

```python
dmelt["age"] = dmelt["group"].str[1:].str.upper()
dmelt.head()
```

```python
dmelt['age'].unique()
```

```python
age_groups = {'04': '0-4', '514': '5-14', '014': '0-14', '1524': '15-24', '2534': '25-34', '3544': '35-44', '4554': '45-54', '5564': '55-64', '65': '+65',
              'U': np.nan}

dmelt['age'] = dmelt['age'].replace(age_groups)
dmelt.head()
```

```python
dmelt = dmelt.drop(columns=["group"])

dmelt.columns
```

```python
col_order = ['country', 'year', 'age', 'gender', 'cases']
dmelt = dmelt[col_order]
dmelt.head()
```

```python
dmelt.to_csv('./myfinal.csv', index=False)
```
