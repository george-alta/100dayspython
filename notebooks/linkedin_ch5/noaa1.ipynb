IV. FORMAT OF "ghcnd-stations.txt"

---

## Variable Columns Type

ID 1-11 Character
LATITUDE 13-20 Real
LONGITUDE 22-30 Real
ELEVATION 32-37 Real
STATE 39-40 Character
NAME 42-71 Character
GSN FLAG 73-75 Character
HCN/CRN FLAG 77-79 Character
WMO ID 81-85 Character

---


```python
urllib.request.urlretrieve(
    "https://www1.ncdc.noaa.gov/pub/data/ghcn/daily/ghcnd-stations.txt", "stations.txt"
)
```

```python
stations = np.genfromtxt(
    "./stations.txt",
    delimiter=[11, 9, 10, 7, 3, 31, 4, 4, 6],
    names=[
        "id",
        "latitude",
        "longitude",
        "eleavtion",
        "state",
        "name",
        "gsn",
        "hcn",
        "wmo",
    ],
    dtype=["U11", "d", "d", "d", "U3", "U31", "U4", "U4", "U6"],
    autostrip=True,
)
```

```python
len(stations)
```

```python
stations
```

```python
pp.plot(stations["longitude"], stations["latitude"], ".", markersize=1)
```

```python
stations_ca = stations[stations["state"] == "CA"]
```
